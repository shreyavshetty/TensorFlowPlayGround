```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data_frame = pd.read_csv("Selfie-dataset/selfie_dataset.txt",sep=" ")
data_frame.columns = ['image_name','popularity_score','partial_faces','is_female','baby','child','teenager','youth','middle_age','senior','white','black','asian','ovaal_face','round_face','heart_face','smiling','mouth_open','frowning','wearing_glasses','wearing_sunglasses','wearing_lipstick','tongue_out','duck_face','black_hair','blond_hair','brown_hair','red_hair','curly_hair','straight_hair','braid_hair','showing_cellphone','using_earphone','using_mirror','braces','wearing_hat','harsh_lighting','dim_lighting']
```

```python
data_frame.head()
```

```python
max_value = np.amax(np.array(data_frame['popularity_score']))
min_value = np.amin(np.array(data_frame['popularity_score']))
    
```

```python
fraction = (max_value-min_value)/3
print(fraction)
class_1 = min_value + fraction
class_2 = class_1 + fraction
class_3 = class_2 + fraction
print(class_1,class_2,class_3)
```

```python
selfie_state = ['Great_Selfie','Average_Selfie','Poor_Selfie']
data_frame['State'] = " "
data_frame.head()
```

```python
list_pop_scores = np.array(data_frame['popularity_score'])

state_list = []

for score_val in list_pop_scores:
    if score_val <= class_1:
        state = selfie_state[0]
    elif score_val > class_1 and score_val <= class_2:
        state = selfie_state[1]
    else:
        state = selfie_state[2]
    state_list.append(state)
    
print(state_list)
```

```python
data_frame['State'] = state_list
```

```python
from sklearn.feature_selection import RFE 
from sklearn.linear_model import LinearRegression 
```

```python
df = great_selfie_df
columns = ['partial_faces','is_female','baby','child','teenager','youth','middle_age','senior','white','black','asian','ovaal_face','round_face','heart_face','smiling','mouth_open','frowning','wearing_glasses','wearing_sunglasses','wearing_lipstick','tongue_out','duck_face','black_hair','blond_hair','brown_hair','red_hair','curly_hair','straight_hair','braid_hair','showing_cellphone','using_earphone','using_mirror','braces','wearing_hat','harsh_lighting','dim_lighting']
score_list = []
for col in columns:
    lm = LinearRegression(fit_intercept=True)
    lm.fit(df[[col]], df['popularity_score'])
    score = lm.score(df[[col]], df['popularity_score'])
    print(score)
    score_list.append(score)
    
print(score_list)
```

```python
score_dict = {}
for row1,row2 in zip(columns,score_list):
    score_dict[row1] = row2

print(score_dict)
```

```python
from collections import OrderedDict
sorted_score_dict = sorted(score_dict, key=score_dict.get, reverse=True)
print(sorted_score_dict)
```

```python
import matplotlib.pyplot as plt
plt.bar(range(len(score_dict)),score_dict.values(),align="center")
plt.xticks(range(len(score_dict)),list(score_dict.keys()) ,rotation="vertical")
plt.rcParams["figure.figsize"] = (15,15)
plt.show()
```
